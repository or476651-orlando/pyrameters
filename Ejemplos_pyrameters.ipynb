```python
from pyrameters.all import pyrameters
```

El primer ejemplo: la gráfica de Petersen, una gráfica de sumo interés en teoría de gráficas.

```python

G = nx.petersen_graph()

resultados = pyrameters(G)

# Desplegamos los parámetros
print("--- Resultados de pyrameters ---")
print(f"Clique Máximo (Nodos): {resultados['maximum_clique']}")
print(f"Número de Clan (ω):     {resultados['clique_number']}")
print(f"Número Cromático (χ):  {resultados['chromatic_number']}")
print(f"Girth (Cintura):       {resultados['girth']}")
```

El siguiente ejemplo es considerado una prueba de estrés, ya que es una gráfica aleatoria. Estas gráficas pertenecen al modelo de Erdös - Renyi. Esta gráfica se define por dos parámetros: n, el número de vértices; y p, una probabilidad (un número entre 0 y 1). Esta es la gráfica tal que cada par de vértices tiene una probabilidad p de estar conectados. Por supuesto, se le puede indicar una semilla (seed) para reproducibilidad.

```python
G = nx.erdos_renyi_graph(20, 0.7, seed = 1984)

resultados = pyrameters(G)

# Desplegamos los parámetros
print("--- Resultados de pyrameters ---")
print(f"Clique Máximo (Nodos): {resultados['maximum_clique']}")
print(f"Número de Clan (ω):     {resultados['clique_number']}")
print(f"Número Cromático (χ):  {resultados['chromatic_number']}")
print(f"Girth (Cintura):       {resultados['girth']}")
```
